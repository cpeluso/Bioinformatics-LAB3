```python
def parse_vcf(filename):
    
    file = list()
    
    with open(filename, "r") as f:
        for line in f:
            if line.startswith("#") == False:
                row    = [x.replace('\n', '') for x in line.split('\t')]
                row[4] = row[4].split(",")
                file.append(row)
                
    return file

def find_SNP(df):
    return df[ 
        df['Reference'].apply(lambda x: len(x)==1) &
            (
                df['Alternative'].apply(lambda x: len(x)==1 and "<*>" not in x and len(x[0]) == 1) |
                df['Alternative'].apply(lambda x: len(x)==2 and "<*>" in x and len(x[0]) == 1) 
            )]

def find_INDEL(df):
    return df[
        df.apply(lambda x: len(x['Reference']) != len(x['Alternative'][0]) and x['Alternative'][0] != "<*>", axis=1)
    ]
```

```python
def parse_gtf(filename):
    
    file = list()
    
    with open(filename, "r") as f:
        for line in f: 
            if line.startswith("#") == False:
                row = [x.replace('\n', '') for x in line.split('\t')]
                file.append(row)
        
    return file

def parse_SAM(filename):
    
    file = list()
    
    with open(filename, "r") as f:
        for line in f: 
            if line.startswith("#") == False:
                row = [x.replace('\n', '') for x in line.split('\t')]
                file.append(row[:11])
        
    return file

def extract_protein_coding_genes(df):
    return df[df.apply(lambda x: x.Feature == "gene" and "protein_coding" in x.Attributes, axis=1)]

def perform_raw_read_count(gtf_df, sam_df):
    
    for gtf_index, gtf_row in gtf_df.iterrows():
        
        for sam_index, sam_row in sam_df.iterrows():
            
            if int(gtf_row.Start) <= int(sam_row.POS) and int(sam_row.POS) + sam_row.LENGTH <= int(gtf_row.End):
                gtf_df.loc[gtf_index, 'Count'] += 1
            
    return gtf_df[gtf_df.Count != 0]
```
